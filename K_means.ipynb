```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/3.01. Country clusters.csv')
```

```python
data.head()
```

```python
new_data = data.iloc[ : , 1:3]
new_data.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(3)
```

```python
kmeans.fit(new_data)
```

```python
prediction = kmeans.fit_predict(new_data)
```
